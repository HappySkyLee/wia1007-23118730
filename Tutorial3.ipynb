```python
n = int(input("Enter the number of row of the 1st array: "));
p = int(input("Enter the number of column of the 1st array or the number of row of the 2nd array: "));
m = int(input("Enter the number of column of the 2nd array: "));
Array_nxp = np.random.rand(n,p);
Array_pxm = np.random.rand(p,m);
dot_product = np.dot(Array_nxp,Array_pxm);
print("5.\n1st array\n",Array_nxp);
print("\n2nd array\n",Array_pxm);
print("\nDot Product:\n",dot_product);
```
